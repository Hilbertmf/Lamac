```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# base
canvas = np.zeros((500, 500), dtype=np.uint8)
canvas[150:350, 150:350] = 255

def translated_img(img, tx, ty):
    rows, cols = img.shape

    #transform matrix
    M = np.float32([[1, 0, tx],
                    [0, 1, ty]])
    result = cv.warpAffine(img, M, (cols, rows))
    return result

moved_right = translated_img(canvas, 120, 0)
moved_diagonal_down = translated_img(canvas, 120, 120)
moved_diagonal_up = translated_img(canvas, 120, -90)

plt.imshow(moved_right, cmap='gray')
plt.show()
plt.imshow(moved_diagonal_down, cmap='gray')
plt.show()
plt.imshow(moved_diagonal_up, cmap='gray')
plt.show()


# def generate_translated_square(dx, dy, size=200, canvas_size=500):
#     canvas = np.zeros((canvas_size, canvas_size), dtype=np.uint8)
#     y_start, y_end = 150 + dy, 350 + dy
#     x_start, x_end = 150 + dx, 350 + dx

#     y_start = max(0, min(canvas_size, y_start))
#     y_end = max(0, min(canvas_size, y_end))
#     x_start = max(0, min(canvas_size, x_start))
#     x_end = max(0, min(canvas_size, x_end))

#     canvas[y_start:y_end, x_start:x_end] = 255
#     return canvas

# sample_centered = generate_translated_square(0, 0)
# sample_up_left = generate_translated_square(-50, -100) 
# sample_down_right = generate_translated_square(100, 80)

# plt.imshow(sample_centered, cmap='gray')
# plt.show()
# plt.imshow(sample_up_left, cmap='gray')
# plt.show()
# plt.imshow(sample_down_right, cmap='gray')
# plt.show()
```
