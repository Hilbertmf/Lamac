```python
import numpy as np
import cv2
import matplotlib.pyplot as plt


# Create image
canvas = np.zeros((500, 500), dtype=np.uint8)
canvas[150:350, 150:350] = 255


# Show image
plt.imshow(canvas, cmap='gray')
plt.show()
cv2.imwrite('binary_square.png', canvas)

# Read image
img = cv2.imread('binary_square.png')
```
