```python
# Learning translation
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# base
canvas = np.zeros((500, 500), dtype=np.uint8)
canvas[150:350, 150:350] = 255

def translated_img(img, tx, ty):
    rows, cols = img.shape

    #transform matrix
    M = np.float32([[1, 0, tx],
                    [0, 1, ty]])
    result = cv.warpAffine(img, M, (cols, rows))
    return result

moved_right = translated_img(canvas, 130, 0)
moved_diagonal_down = translated_img(canvas, 140, 140)
moved_diagonal_left_up = translated_img(canvas, -140, -140)

plt.imshow(moved_right, cmap='gray')
plt.show()
plt.imshow(moved_diagonal_down, cmap='gray')
plt.show()
plt.imshow(moved_diagonal_left_up, cmap='gray')
plt.show()

```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import random

# rotate 
squares = generate_random_squares(50, 400)
square = squares[0]


def rotate_squares(squares):
    for i, square in enumerate(squares):
        rnd = random.random()
        if rnd > 0.5:
            # use image moment to locate square center
            m = cv.moments(square)
            if m['m00'] == 0:
                continue
            # calculate centroid
            cx = int(m['m10'] / m['m00'])
            cy = int(m['m01'] / m['m00'])
            center = (cx, cy)
            angle_deg = random.randint(15, 45)
            rot_matrix = cv.getRotationMatrix2D(center, angle_deg, 1.0)
            rotated_sqr = cv.warpAffine(square, rot_matrix, square.shape)
            squares[i] = rotated_sqr
            
rotate_squares(squares)
for i in range(3):
    plt.imshow(squares[i], cmap='gray')
    plt.show()
```
